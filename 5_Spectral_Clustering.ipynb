# Clusterizaci칩n de trayectorias usando Clustering Espectral y KMeans

## Dependencias

````python
import urllib
import zipfile
import os
import scipy.io
import math

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.spatial.distance import directed_hausdorff
from scipy import linalg

from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans

# Some visualization stuff, not so important
sns.set()
plt.rcParams['figure.figsize'] = (12, 12)

````

## Funciones

````python
def plot_cluster(traj_lst, cluster_lst):
    cluster_count = np.max(cluster_lst) + 1
    
    for traj, cluster in zip(traj_lst, cluster_lst):
        plt.plot(traj[:, 0], traj[:, 1], c=plt.cm.tab20(cluster))
    plt.show()

````

## Obtenci칩n de los datos

Trabajaremos con el dataset [LABOMNI](http://cvrr.ucsd.edu/bmorris/datasets/dataset_trajectory_clustering.html) que contiene trayectorias de humanos caminando por un laboratorio capturadas usando una c치mara omni-direccional.

````python

dataset_link = 'http://cvrr.ucsd.edu/LISA/Datasets/TrajectoryClustering/CVRR_dataset_trajectory_clustering.zip'
data_folder = 'data'
filename = '%s/labomni.mat' % data_folder

# Download the dataset
is_download_required = not os.path.exists(data_folder)

if is_download_required:
    zip_filename = 'data.zip'
    urllib.request.urlretrieve(dataset_link, zip_filename)
    zip_ref = zipfile.ZipFile(zip_filename, 'r')
    zip_ref.extractall(data_folder)
    zip_ref.close()

# Import dataset
traj_data = scipy.io.loadmat(filename)['tracks']

traj_lst = []
for data_instance in traj_data:
    traj_lst.append(np.vstack(data_instance[0]).T)

````

## Matriz de similitud/distancia/adyacencia

Para calcular la similitud entre dos trayectorias, ocuparemos la distancia de Hausdorff programada en [```scipy.spatial.distance.directed_hausdorff```](https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.directed_hausdorff.html). Esto crear치 la matriz ```A```.

````python
%%time
def hausdorff( u, v):
    d = max(directed_hausdorff(u, v)[0], directed_hausdorff(v, u)[0])
    return d

traj_count = len(traj_lst)
A = np.zeros((traj_count, traj_count))

# This may take a while
for i in range(traj_count):
    for j in range(i + 1, traj_count):
        distance = hausdorff(traj_lst[i], traj_lst[j])
        A[i, j] = distance
        A[j, i] = distance
````

````python
print(A.shape)
A[:10, :4]
````

## Clustering Espectral

Obtenemos la $D$ que es una matriz diagonal en la que tendremos el ```grado``` de la trayector칤a $i$ en la $i-칠sima$ entrada.

$D_{i} = \sum_j A_{i, j}$

````python
D_vect = A.sum(axis = 1)
D = np.diag(D_vect)
D[:10, :4]
````

Calculamos la matriz $L$. No es la matriz Laplaciana de $A$, pero esta relacionada.

$L = D^{-1/2}~A~D^{-1/2}$

````python
D_sqrtinv = np.diag(np.sqrt(1/D_vect))
L_norm = D_sqrtinv @ A @ D_sqrtinv
L_norm[:10, :3]
````

Hacemos la descomposici칩n en valores singulares 

$L = U~\Sigma~V^T$.

Los eigenvectores van son las columnas de $U$.

````python
U, sigma, _ = linalg.svd(L_norm, full_matrices=False)
print(U.shape, sigma.shape)
sigma[:10]
````

Definimos $k$ como el n칰mero de clusters que queremos crear y nos *quedamos* con los $k$ eigenvectores correspondientes a los $k$ eigenvalores m치s grandes.

````python
k = 10
Uk = U[:, :k]
Uk.shape
````

Finalmente, normalizamos cada fila de la matriz resultante, $U_k$, y ocupamos 칠stas como *feature vectors* para KMeans.

````python
y_pred = KMeans(n_clusters = k).fit_predict(normalize(Uk))
y_pred
````

````python
plot_cluster(traj_lst, y_pred)
````

````python
cluster_n = 5
plot_cluster([i for (i, v) in zip(traj_lst, y_pred == cluster_n) if v], y_pred[y_pred == cluster_n])
````
